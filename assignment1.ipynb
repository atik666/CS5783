Perform the following operations:

1. For each pixel in the image, i.e., every element in the array img,
compute the Euclidean distance to the following points [255, 0, 0], [0,
255, 0] and [0, 0, 255]. Once computed, set the value of the pixel to
the point with the least distance. Save the image as output1.png

2. Set the pixels that fall within a square (whose sides are of length 50)
at the center of the image to be [0,0,0]. Save the image as
output2.png.

#### Import files

```python
import cv2
import numpy as np
import copy
import matplotlib.pyplot as plt
```

#### Read images

```python
img = cv2.imread('test.png')
img_new1 = copy.copy(img)
img_new2 = copy.copy(img)

plt.imshow(img)
plt.show()
```

#### Calculate Euclidean distance and replace the values

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        a = img[i,j,:]
        a1 = np.linalg.norm(a - [255,0, 0])
        b1 = np.linalg.norm(a - [0,255, 0])
        c1 = np.linalg.norm(a - [0,0, 255])
        
        index_min = np.argmin([a1,b1,c1])
        
        if index_min == 0:
            img_new1[i,j,:] = [255,0,0]
        elif index_min == 1:
            img_new1[i,j,:] = [0,255,0]
        elif index_min == 2:
            img_new1[i,j,:] = [0,0,255]
```

#### Show output image: output1.png

```python
plt.imshow(img_new1)
plt.title("output1.png")
plt.savefig("output1.png")
plt.show()
```

#### Find square in the image

```python
img_new2[int(img_new2.shape[0]/2 - 25): int(img_new2.shape[0]/2) + 25,
    int(img_new2.shape[1]/2 - 25): int(img_new2.shape[1]/2) + 25, :] = [0,0,0] # replacing with [0,0,0]
```

#### Show output image: output2.png

```python
plt.imshow(img_new2)
plt.title("output2.png")
plt.savefig("output2.png")
plt.show() 
```
