```python
import numpy as np
```

```python
# Define activation function
def sigmoid(z):
	return 1/(1 + np.exp(-z))

def sigmoid_derivative(z):
  return sigmoid(z)*(1-sigmoid(z))
```

```python
# Define input
x = np.random.uniform(0,1,(2,1))
y = 0

print('X:', x,"\n", "y:", y)
```

```python
# Define initial parameters
W1 = np.random.rand(1,2)
b1 = np.random.rand(1,1)

print("Initial parameters:")
print("W1", W1)
print("b1", b1)
```

```python
# Fwd Pass
z1 = np.dot(W1, x) + b1
a1 = sigmoid(z1)

# Initial output
print("Prediction:", a1, "Actual:", y)
```

```python
loss = 'mse' # 'log'
# Backward pass
# Compute gradients
if loss == 'mse':
    dz1 = np.square(np.subtract(y,a1)).mean()
elif loss == 'log':
    dz1 = -1 * np.mean(y * np.log(a1) + (1 - y) * np.log(1-a1))
db1 = dz1
dW1 = np.dot(dz1, a1.T)

# Update weights
W1 = W1 - dW1
b1 = b1 - db1
```

```python
# Fwd Pass with new weights
z1New = np.dot(W1, x) + b1
a1New = sigmoid(z1New)

print("Initial Prediction", a1, "New Prediction:", a1New, "Actual:", y)
```

```python
# Backprop
def backprop(W1, b1, loss):
    for i in range(25):
        if loss == 'mse':
            dz1 = np.square(np.subtract(y,a1)).mean()
        elif loss == 'log':
            dz1 = -1 * np.mean(y * np.log(a1) + (1 - y) * np.log(1-a1))

        dW1 = np.dot(dz1, a1.T)
        db1 = dz1
    
        W1 = W1 - dW1
        b1 = b1 - db1
        z1New = np.dot(W1, x) + b1
        a1New = sigmoid(z1New)
        
        print("New prediction: %f" %a1New, "old prediction: %f" %a1)
```

```python
backprop(W1, b1, loss = 'mse')
```

```python
backprop(W1, b1, loss = 'log')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MSE(y, y_hat):
   return np.square(np.subtract(y,y_hat)).mean()
def Log(y, y_hat):
    return -1 * np.mean(y * np.log(y_hat) + (1 - y) * np.log(1-y_hat))
def sigmoid(z):
    return 1/(1 + np.exp(-z))
def dEdW_1 (y, y_hat,x_1):
   return np.sum(-(y - y_hat) * x_1)
def dEdW_0 (y, y_hat):
   return np.sum(-(y - y_hat))
```

```python
N = 200
slope = 3
mu, sigma = 0, 0.1 # mean and standard deviation
intercept = np.random.normal(mu, sigma, N)
x_1 = np.random.rand(N)
y = slope*x_1 + intercept
```

```python
def regression(x_1, y, loss = 'MSE'):
    w_1 = np.random.uniform(-2,-3,1)
    w_0 = np.random.uniform(0,2,1)
    x_0 = 1
    y_hat = w_1*x_1 + w_0*x_0
    epoch = 300
    eta = 0.001
    E = []
    for ep in range(epoch):
        Z = w_1 * x_1 + w_0*x_0
        y_hat = sigmoid(Z)
        if loss == 'MSE':
            error = MSE(y, y_hat)
        E.append(error)
        dEdw_1 = dEdW_1(y, y_hat, x_1)
        dEdw_0 = dEdW_0(y, y_hat)
        w_1 = w_1 - eta*dEdw_1
        w_0 = w_0 - eta*dEdw_0
        if ep % 20 == 0:
           plt.scatter(x_1, y, s=np.pi*10, c='r', alpha=0.5)
           plt.plot(x_1, y_hat, '-b', label="MSE= %.2f" %error + '\n' + "Epoch Number= %d" %(ep))
           plt.xlabel('x (Our Data)')
           plt.ylabel('y_hat (Network Outputs)')
           plt.legend()
           plt.ylim(-3, 5)
           plt.grid()
           plt.show()
```

```python
w_1 = np.random.uniform(-2,-3,1)
w_0 = np.random.uniform(0,2,1)
x_0 = 1
y_hat = w_1*x_1 + w_0*x_0
epoch = 300
eta = 0.001
E = []
for ep in range(epoch):
    Z = w_1 * x_1 + w_0*x_0
    y_hat = sigmoid(Z)
    error = MSE(y, y_hat)
    E.append(error)
    dEdw_1 = dEdW_1(y, y_hat, x_1)
    dEdw_0 = dEdW_0(y, y_hat)
    w_1 = w_1 - eta*dEdw_1
    w_0 = w_0 - eta*dEdw_0
    if ep % 20 == 0:
       plt.scatter(x_1, y, s=np.pi*10, c='r', alpha=0.5)
       plt.plot(x_1, y_hat, '-b', label="MSE= %.2f" %error + '\n' + "Epoch Number= %d" %(ep))
       plt.xlabel('x (Our Data)')
       plt.ylabel('y_hat (Network Outputs)')
       plt.legend()
       plt.ylim(-3, 5)
       plt.grid()
       plt.show()
```

```python
plt.plot(E, 'r')
plt.grid()
plt.title("Error")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.show()
```
